```
import pandas as pd

column_names = ["Pair","-2A","-1A","0A","+1A","+2A","-2R","-1R","0R","+1R","+2R","label"]
Dataset = pd.DataFrame(columns = column_names)

directory_in_str = '/content/RNADataClasses'
directory = os.fsencode(directory_in_str)

for file in os.listdir(directory):
  filename = os.fsdecode(file)
  
#read File and preprocess it for later Function to work on
s = pd.read_csv('/content/RNADataClasses/{}'.format(filename),header=None)
s.drop(0,axis=0,inplace=True)
s.drop(0,axis=1,inplace=True)
s.iloc[0,0] = 'Amino'
s.columns = s.iloc[0]
s.drop(1,axis=0,inplace=True)
s.reset_index(drop=True,inplace=True)
coloumns = s.columns[1:]
rows = s['Amino']

#loop over each file to convert it to appropiate Dataset Format
for row in s.itertuples(index=True):
  index = 0
  row_index = row[0]
  for value in row[2:]:
    x = np.array([1000, 1000, 1000, 1000, 1000, 1000, 1000, 1000, 1000, 1000, 1000, 's'])
    #get pair information
    rna = coloumns[index]
    amino = row[1]
    pair = "({},{})".format(rna,amino)
    x[0] = pairs_dictionary[pair]
    
    #current amino acid
    x[3] = Amino_dictionary[amino]
    #current rna
    x[8] = RNA_dictionary[rna]
    
    #get amino acid information
    if (row_index >= 2):
      x[1] = Amino_dictionary[rows[row_index - 1]]
      x[2] = Amino_dictionary[rows[row_index - 2]]
    
    if (row_index <= len(rows) - 3):
      x[4] = Amino_dictionary[rows[row_index + 1]]
      x[5] = Amino_dictionary[rows[row_index + 2]]
    
    #get Rna information
    if (index >= 2):
      x[6] = RNA_dictionary[coloumns[index - 1]]
      x[7] = RNA_dictionary[coloumns[index - 2]]
    if (index <= len(coloumns) - 3):
      x[9] = RNA_dictionary[coloumns[index + 1]]
      x[10] = RNA_dictionary[coloumns[index + 2]]
    #get label information
    x[11] = value
    #create and appaned DataFrame
    df = pd.DataFrame([x], columns=["Pair","-2A","-1A","0A","+1A","+2A","-2R","-1R","0R","+1R","+2R","label"])
    Dataset = Dataset.append(df,ignore_index=True)
    #increment index
    index += 1
```

#**Model**

```
import tensorflow as tf
from sklearn.datasets import make_circles
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD,Adam
from matplotlib import pyplot
from sklearn.model_selection import train_test_split

#selct x and y 
x = Dataset.iloc[:,:-1].astype(int)
y = Dataset.iloc[:,-1].astype(str)
Y = label_encoder.fit_transform(y)
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(x, Y, test_size=0.2, random_state=1)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=1)
# define model
model = Sequential()
model.add(Dense(50,activation='relu',input_shape=(X_train.shape[1],), kernel_initializer='he_uniform'))
model.add(Dense(25, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(9, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1, activation='softmax'))
opt = Adam(lr=0.01)
#compile model
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
# fit model
history = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=20, batch_size=64)
```
